### 지니계수 계산 방식 - 높을수록 불순도 높다.
##### 불순도(불순한 정도) 확인
- 특정 사건이 일어날 확률을 yes_prob이라 하고
- 특정 사건이 일어나지 않을 확률을 no_prob이라 했을 때
- yes_prob * (1 - yes_prob) 을 A라 하고
- no_prob * (1 - no_prob) 을 B라 하면
- A + B 가 지니계수다.

- 생각해보니 일어날 확률과 일어나지 않을 확률을 서로 곱하고 2를 곱한 값이네 ㅎㅎ

### 배깅 - bagging = bag + ing -> bag = bootstrap + aggregation
##### bootstrap으로 두출한 결과가 여러개인데 이를 합산.
- bootstrap : 데이터의 row를 무작위로 추출하여 새로운 데이터 셋 생성(복원추출)
- 같은 데이터로 같은 알고리즘(랜덤성X)에 넣고 학습하면 항상 같은 결과 나옴
- DecisionTree는 참고로 랜덤성이 없음. 
- bootstrap을 이용하면 데이터 row를 추출해 새로운 데이터 셋을 생성, 예를 들어 첫번째 데이터 셋을 만들고 두번째 데이터 셋을 생성했을 때, 
- 랜덤으로 뽑는 과정에서 중복되는 row가 있지만 기본적으로 구성이 다르다. (어떤 row는 여기에는 있고 저기에는 없고 한곳에 중복으로 들어가있을수도 있음)
- 데이터 셋이 있으니 학습을 시킬 수 있고, 구성이 다르기 때문에 랜덤성이 없는 같은 알고리즘에 넣어서 학습을 해도 같은 결과가 나오지 않는다.

- 정리하면, bootstrap 이용하면 새로운 데이터 셋 만들 때마다 row의 구성이 조금씩 다르게 됨.
- 서로 조금씩 다른 데이터 셋을 bootstrap으로 n개 만들고 
- n개의 데이터 셋에 대하여 알고리즘 학습 진행(서로 조금씩 다르게 학습됨)
- n개의 서로 조금씩 다르게 학습된 알고리즘들의 예측 결과를 모아모아서 합산(aggregation)
- 예) RandomForest의 경우 대표적인 배깅 알고리즘입니다. 
- 200개(옵션설정)의 서로 다른 데이터 셋을 bootstrap으로 만들고 decision tree 학습
- 그러면 서로 조금씩 다른 decision tree 들이 200개 학습될 것입니다.
- 예측값 도출시 학습된 200개의 알고리즘들의 결과값을 합산(aggregation) 합니다.
- RandomForest 의 Random은 row 무작위성과 featue 무작위 선택이 포함
- https://passionpython.tistory.com/61

### 부스팅 - boost + ing -> 상승. 밀어올리다(정확도를...?)

##### 혼동행렬 구성용소 알아보기
- T / F -> 맞춘 것이 T(True), 틀린 것이 F(False)
- P / N -> 예측 결과가 1이면 P(Positive), 0이면 N(Negative)
- T / F 중 하나 뽑고, P / N 중 하나 뽑아서 조립한 것.



##### 분류 문제의 평가지표들 : 
- accuracy -> 전체 중에 몇 퍼센트 맞춤?
- acc = (TP + TN) / (TP + TN + FN + FP)
- precision -> 모델이 1로 예측한 것 전체 중에서 맞춘 비율
- prec = TP / (TP + FP)
- recall -> 실제 값이 1인 것 중 맞춘 비율
- recall = TP / (TP + FN)
- fall-out -> 실제 값이 0인 것 중 틀린 비율(ROC_Curve 그래프 용)
- fall-out = FP / (FP + TN)

인디언 기우제 precision으로 보면 개판, recall로 보면 높다

##### 언제 accuracy 대신 precision, recall 등이 고려될까?
- class-imbalance issue 발생시 accuracy 의미가 떨어짐
- 클래스 불균형 문제 -> 경우의 수가 한 쪽으로 치우진 경우.


##### FP 와  FN 해석하기 - precision / recall 은 어떻게 다른 것일까?
- 화재경보(불 나면 1, 안 나면 0)
- FP - 불이 났다고 판단했으나 불 안남 = False Alarm
- FN - 불이 안났다고 판단했으나 불 남 = Miss 
- 스팸메일 분류(스팸이면 1, 아니면 0)
- FP - 스팸이라고 판단했으나 스팸 아님 = False Alarm
- FN - 스팸 아니라고 판단했으나 스팸임 = Miss
- FP - FN 관계는 Trade-off

```python
from sklearn.metrics import confusion_matrix as cm
TN, FP, FN, TP = cm(for_plot['actual'], for_plot['predict']).ravel()
print(TN, FP, FN, TP)
```

```python
acc = (TP + TN) / (TP + TN + FN + FP)
prec = TP / (TP + FP)
recall = TP / (TP + FN)
fall = FP / (FP + TN)
print(acc,prec,recall)
```

### ROC_Curve가 어떻게 그려지는가?
##### 확신의 정도와 실제 정답을 이용하여 y축을 recall, x축을 fall-out 좌표로 사용
- recall -> 원래 1인것들 중 맞춘 비율
- fall-out -> 1- 원래 0인것들 중 맞춘 비율
- 다 1이라 예측하면? -> recall 1.0, fall-out 1.0 -> (1,1) 좌표
- 다 0이라 예측하면? -> recall 0.0, fall-out 0.0 -> (0,0) 좌표 
- 판단 기준을 0 ~ 100% 까지 조금씩 바꿔가며 그때마다 confusion-matrix 획득
- confusion-matrix로부터 평가지표 값 산출
- 만약 판단 기준을, 0, 10, 20, ..... 100% 으로 11번 다르게 한 경우?
- confusion-matrix도 11개 나오고 좌표평면에 점도 11개 찍을 수 있음.
- 이 점들을 이어 그리면 ROC_Curve 완성.

판단기준이란 예를 들어 0.5라 하면 pos 예상확률이 0.5를 넘으면 pos로 예상하고 넘지 못하면 neg로 가는것

```python
# 같은 1 이라는 결과를 내더라도, row마다 확신의 정도가 다릅니다.
r2['actual'] = y_test
r2
```

```python
threshold = 0.1 # threshold 값 0~1 사이 변화에 따라 FN - FP 조율 가능
r2['predict'] = 0
r2.loc[r2['pos'] > threshold, 'predict'] = 1
TN, FP, FN, TP = cm(r2['actual'], r2['predict']).ravel()
acc = (TP + TN) / (TP + TN + FN + FP)
prec = TP / (TP + FP)
recall = TP / (TP + FN)
fall = FP / (FP + TN)
print(FP, FN)
print(acc,prec,recall)
```

### ROC_Curve 정리
##### 확신의 정도를 이용
- 학습 후에는 예측시 알고리즘 내부적으로 확신의 정도(Confidence)가 있습니다.
- 예측력이 좋은 알고리즘이라면 맞는 답에 대한 확신의 정도가 클 것이라 판단.

```python
data_points = 100
score_list = []
for i in range(data_points):
    threshold = i / data_points
    r2['predict'] = 0
    r2.loc[r2['pos'] > threshold, 'predict'] = 1
    TN, FP, FN, TP = cm(r2['actual'], r2['predict']).ravel()
    acc = (TP + TN) / (TP + TN + FN + FP)
    prec = TP / (TP + FP)
    recall = TP / (TP + FN)
    fall = FP / (FP + TN)
    scores = {'precision': prec, 'recall': recall, 'fall' : fall}
    score_list.append(scores)
curve_df = pd.DataFrame(score_list)
sns.lineplot(data=curve_df, x='fall', y='recall')
```

### Hyperparameter - Tuning
##### 초매개변수 조정, 하이파라미터 튜닝
- 알고리즘의 옵션을 조정하여 성능을 튜닝하는 방법
- 옵션에 따라 성능이 달라집니다.
- 최적의 옵션값을 찾아 알고리즘 성능을 극대화 시키는 접근법.


```python
model = rf(max_depth=10, max_features=0.1)
# 옵션 값이 달라지면, 평가지표 수치가 달라집니다.
model.fit(X_train, y_train)
for_plot = pd.DataFrame()
for_plot['predict'] = model.predict(X_test)
for_plot['actual'] = y_test
TN, FP, FN, TP = cm(for_plot['actual'], for_plot['predict']).ravel()
recall = TP / (TP + FN)
print(recall)
```

```python
def scorer(model):
    model.fit(X_train, y_train)
    for_plot = pd.DataFrame()
    for_plot['predict'] = model.predict(X_test)
    for_plot['actual'] = y_test
    TN, FP, FN, TP = cm(for_plot['actual'], for_plot['predict']).ravel()
    recall = TP / (TP + FN)
    return recall
```

```python
# scorer 함수에 옵션을 설정한 모델을 넣으면, 평가지표 값이 결과로 나오게 설정
model = rf(max_depth=7, max_features=0.6)
scorer(model)
```
